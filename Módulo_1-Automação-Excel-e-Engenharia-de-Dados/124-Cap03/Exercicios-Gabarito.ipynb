Então considere o dataframe abaixo e crie o código Python que gera as 5 tabelas pivot a seguir. Compare suas respostas com o gabarito fornecido.

```python
# Dataframe
df = pd.DataFrame({
    'Montadora': ['Porsche', 'Honda', 'Toyota', 'BMW', 'Porsche', 'Honda', 'Toyota', 'BMW'],
    'Localidade': ['EUA', 'EUA', 'EUA', 'EUA', 'Alemanha', 'Alemanha', 'Alemanha', 'Alemanha'],
    'Modelo':['Taycan', 'Civic', 'Corolla', 'Z4', 'Macan', 'Accord', 'Camry', 'X3'],
    'Unidades_Vendidas': [97, 75, 84, 92, 120, 98, 76, 81]})
```

```python
df
```

```python
# Tabela Pivot 1
pd.pivot(data = df, 
         index = 'Montadora', 
         columns = 'Localidade', 
         values = ['Unidades_Vendidas']).fillna(0)
```

```python
# Tabela Pivot 2
df.pivot(columns = 'Localidade', 
         values = ['Modelo', 'Unidades_Vendidas'], 
         index = 'Montadora')
```

```python
# Tabela Pivot 3
df.pivot(columns = 'Localidade', 
         values = ['Modelo'], 
         index = 'Montadora')
```

```python
# Tabela Pivot 4
df.pivot_table(index = 'Montadora', 
               columns = 'Localidade', 
               values = ['Modelo', 'Unidades_Vendidas'],
               aggfunc = {'Modelo':'count', 'Unidades_Vendidas':'mean'})
```

```python
# Tabela Pivot 5
df.pivot_table(index = ['Montadora', 'Modelo'], 
               columns = 'Localidade', 
               values = ['Unidades_Vendidas'],
               aggfunc = {'Unidades_Vendidas':'mean'}).fillna(0)
```
